```r
library(shiny)
library(readxl)
library(ggplot2)
library(scales)
library(dplyr)
library(tidyr)
library(shinythemes)

dataset <- read_excel('Dataset.xlsx')
US <-  read_excel('US.xlsx')
NE <-  read_excel('NE.xlsx')
S <-  read_excel('S.xlsx')
MW <-  read_excel('MW.xlsx')
W <-  read_excel('W.xlsx')

ui <- fluidPage(theme=shinytheme("spacelab"),
    titlePanel("Historical Mortgage Affordability Dashboard"),
    tabsetPanel(
        tabPanel("Plots",
                 fluidRow(
                     column(4, selectInput("select_region",
                                          label= "Region:",
                                          #choices= list("US Total" = "US", 
                                                        #"Northeast" = "NE", 
                                                        #"South" = "S", 
                                                        #"Midwest" = "MW", 
                                                        #"West" = "W"),
                                          #selected= "US")),
                                          choices= c("US Total", "Northeast", "South", "Midwest", "West"),
                                          selected= "US Total")),
                     column(8, sliderInput("select_year",
                                       label= "Year:",
                                       min=1978, max=2017, value=2008, sep="", width="100%"))
                         ),
                 fluidRow(
                     column(3, plotOutput("plot1", height="390px")),
                     column(3, plotOutput("plot2", height="390px")),
                     column(3, plotOutput("plot3", height="390px")),
                     column(3, plotOutput("plot4", height="390px"))
                         ),
                 fluidRow(
                     column(3, htmlOutput("text1")),
                     column(3, htmlOutput("text2")),
                     column(3, htmlOutput("text3")),
                     column(3, htmlOutput("text4"))
                         )
                ),
        tabPanel("Sources", 
                 fluidRow(
                     column(12, htmlOutput("source_text"))
                 )
                ),
        tabPanel("Methodology",
                 fluidRow(
                     column(12, htmlOutput("methodolgy_text"))
                 )
                ),
        tabPanel("Data", dataTableOutput("table"))    
                )
        )

server <- function(input, output) {
    
    #df <- ifelse(input$select_region == "W", W, ifelse(input$select_region == "MW", MW, ifelse(
    #    input$select_region == "S", S, ifelse(input$select_region == "NE", NE, US))))
    
    datasetInput <- reactive({switch(input$select_region, 
                                     "US Total" = US,
                                     "Northeast" = NE,
                                     "South" = S,
                                     "Midwest" = MW,
                                     "West" = W)})
  
    datasetInputyear <- reactive({subset(datasetInput(), Year==input$select_year)})
    
    output$plot4 <- renderPlot({
        #p4<- ggplot(data = gather(subset(datasetInput(), Year==input$select_year), Race, AFI, T_AFI:H_AFI, factor_key=TRUE)) + 
         p4<- ggplot(data = gather(datasetInputyear(), Race, AFI, T_AFI:H_AFI, factor_key=TRUE)) +   
            theme_bw() +
            ggtitle("Affordability Index") +
            scale_y_continuous(name="Index Score", breaks= seq(0,2,0.25), limits=c(0,2)) +
            scale_x_discrete(name="Race/Ethnicity", 
                                labels=c("T_AFI"="All","W_AFI"="White","B_AFI"="Black","H_AFI"="Hispanic")) +
            theme(legend.position="none") +
            geom_bar(aes(x=Race, y=AFI, fill=Race, width=.8), stat="identity") +
            geom_hline(aes(yintercept=1, color="red"), size=1.25) +
            scale_fill_manual(values=c("lightslategray","firebrick","navy","forestgreen")) +
            annotate("text", x="W_AFI", y=2, label="A score of 1 means the median") +
            annotate("text", x="W_AFI", y=1.875, label="household makes exactly") +
            annotate("text", x="W_AFI", y=1.75, label="enough income to qualify") +
            annotate("text", x="W_AFI", y=1.625, label="for the median mortgage")
        print(p4)
    })
    
    output$plot1 <- renderPlot({
        p1<- ggplot(datasetInput(), aes(x=Year)) + 
            theme_bw() +
            ggtitle("Costs") +
            scale_y_continuous(name="Dollars", breaks= seq(0,425000,25000), limits=c(5000,425000)) +
            theme(legend.position="none") +
            geom_line(aes(y=Med_PP, color="PP"), size=1.02) + 
            geom_line(aes(y=Med_Mtg, color="Mtg"), size=1.02) + 
            geom_line(aes(y=Med_DP, color="DP"), size=1.02) + 
            geom_vline(aes(xintercept=input$select_year, alpha =.5)) +
            scale_color_manual(values= c(PP="darkmagenta",Mtg="mediumpurple",DP="orchid"))
        print(p1)
    })
    
    output$plot2 <- renderPlot({
        p2<- ggplot(datasetInput(), aes(x=Year)) + 
            theme_bw() +
            ggtitle("Rates") +
            scale_y_continuous(name="Percent", breaks= seq(0,20,.5), limits=c(3.5, 16.0)) +
            theme(legend.position="none") +
            geom_line(aes(y=Med_Rate, color="Rate"), size=1.02) + 
            geom_line(aes(y=Med_APR, color="APR"), size=1.02) + 
            geom_vline(aes(xintercept=input$select_year, alpha =.5))+
            scale_color_manual(values= c(Rate="peru",APR="gold"))
        print(p2)
    })
    
    output$plot3 <- renderPlot({
        p3<- ggplot(datasetInput(), aes(x=Year)) + 
            theme_bw() +
            ggtitle("Incomes") +
            scale_y_continuous(name="Dollars", breaks= seq(0,75000,5000), limits=c(5000,75000)) +
            theme(legend.position="none") +
            geom_line(aes(y=Qinc, color="Qinc"), size=1.5) +
            geom_line(aes(y=T_Med_Inc, color="Med_Inc"), size=1.02) +  
            geom_line(aes(y=W_Med_Inc, color="W_Med_Inc"), size=1.02) +
            geom_line(aes(y=B_Med_Inc, color="B_Med_Inc"), size=1.02) +
            geom_line(aes(y=H_Med_Inc, color="H_Med_Inc"), size=1.02) +
            geom_vline(aes(xintercept=input$select_year, alpha =.5)) +
            scale_color_manual(values= 
                       c(Med_Inc="lightslategray",Qinc="turquoise",W_Med_Inc="firebrick",B_Med_Inc="navy",H_Med_Inc="forestgreen"))
        print(p3)
    })
    
    output$text1 <- renderText({
        {HTML("The median home sales price in", input$select_year,"was <font color=\"DarkMagenta\"><b>", 
             datasetInputyear()$Med_PP,"</font></b><br/>The median mortgage amount was <font color=\"MediumPurple\"><b>", 
             datasetInputyear()$Med_Mtg,"</font></b><br/>The median down payment amount was <font color=\"Orchid\"><b>",
             datasetInputyear()$Med_DP,"</font></b><br/>The median down payment percentage was<font color=\"Black\"><b>",
             datasetInputyear()$Med_DP_Per,"</font></b><br/>The median monthly mortgage payment was<font color=\"Black\"><b>",
             datasetInputyear()$Pymt)} 
    })
    
    output$text2 <- renderText({
        {HTML("The median contract interest rate in", input$select_year,"was <font color=\"Peru\"><b>",
             datasetInputyear()$Med_Rate,"</font></b><br/> The median APR was <font color=\"Gold\"><b>",
             datasetInputyear()$Med_APR)}
    })
    
    output$text3 <- renderText({
        {HTML("The annual household income needed to qualify for the median mortgage amount in", input$select_year,
             "was <font color=\"Turquoise\"><b>", datasetInputyear()$Qinc,
             "</font></b><br/>The median household income was<font color=\"LightSlateGray\"><b>",
             datasetInputyear()$T_Med_Inc,"</font></b><br/>The median white household income was<font color=\"Firebrick\"><b>",
             datasetInputyear()$W_Med_Inc,"</font></b><br/>The median black household income was<font color=\"Navy\"><b>",
             datasetInputyear()$B_Med_Inc,"</font></b><br/>The median hispanic household income was<font color=\"ForestGreen\"><b>",
             datasetInputyear()$H_Med_Inc)}
    })
    
    output$text4 <- renderText({
        {HTML("The overall household score in", input$select_year,"was<font color=\"LightSlateGray\"><b>",
             datasetInputyear()$T_AFI,"</font></b><br/>The white household score was<font color=\"Firebrick\"><b>",
             datasetInputyear()$W_AFI,"</font></b><br/>The black household score was<font color=\"Navy\"><b>",
             datasetInputyear()$B_AFI,"</font></b><br/>The hispanic household score was<font color=\"ForestGreen\"><b>",
             datasetInputyear()$H_AFI)}
    })   
    
    output$source_text <- renderText({ 
        {HTML("<br><b><font color=\"Black\">Federal Housing Finance Agency data</b></font>
                    https://www.fhfa.gov/DataTools/Downloads/Pages/Monthly-Interest-Rate-Data.aspx<br>
                <br>Historical mortgage data from the FHFA's Monthly Interest Rate Survey, specifically the 
                <b>By State - Table15</b> found under <b>Historical Summary Tables</b>, were selected for use. 
                The dataset consists of average annual figures for several terms and conditions of single-family, 
                conventional loans, reported by State. The data is provided by a sample of savings institutions, 
                commercial banks, and mortgage loan companies across the country.<br><br>
               <br><b><font color=\"Black\">United States Census Bureau data</b></font>
                    https://www.census.gov/data/tables/time-series/demo/income-poverty/historical-income-households.html<br>
                <br>Historical income data collected by the Census Bureau from several national surveys, specifically the
                <b>Table H-6. Regions-by Median and Mean Income</b> datasets, were selected for use. The datasets consist
                of annual median and mean income figures for Census Bureau defined regions and races/ethnicities.")}
    })
    
    output$methodolgy_text <- renderText({
        {HTML("<br><b><font color=\"Black\">Master Dataset</b></font>
                <br>The master dataset found under the <b>Data</b> tab of this dashboard was created by combining the datasets
                mentioned under <b>Sources</b> and calculating several new variables based on the available data.<br>
                <br>The FHFA data was condensed from States into Regions as defined by the Census Bureau
                (https://www.census.gov/geo/reference/webatlas/regions.html). The median value per Region for each variable 
                was utilized for each year.<br>
                <br>Only the following racial/ethnic data from the Census Bureau datatables were used: <b>all races, white alone,
                white, black alone, black, hispanic(any race)</b>. The decision was made to only utilize these data in order
                to minimize the potential double counting of households identifying under more than one category. The Asian
                dataset was excluded entirely because data collection only began in 2002.<br><br>
               <br><b><font color=\"Black\">New Computations</b></font>
                <br>Mortgage Amounts were calculated by multiplying the Purchase Prices by the Loan-to-Price Ratio percentages.
                <br>Down Payment Amounts were calculated by subtracting the Mortgage Amounts from the Purchase Prices.
                <br>Down Payment Percentages were calculated by dividing the Down Payment Amounts by Purchase Prices.<br>
                <br>Monthly Payment Amounts were calculated as follows (this is equivalent to the PMT() function in Excel): 
                <b>payment = (interest_rate/(1-(1+interest_rate)^-loan_term))*-median_mortgage_amount</b><br>
                <br>Qualifying Incomes were calculated as follows: <b>qualifying_income = payment/0.28*12</b><br>
                <br>The Mortgage Affordability Index was calculated as follows:
                <b> mai = (median_income/qualifying_income)</b>")}
    })
    
    output$table <- renderDataTable(dataset)
}

shinyApp(ui=ui, server=server)
```
